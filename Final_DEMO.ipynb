```python
import gradio as gr
import numpy as np
from PIL import Image, ImageDraw
from tensorflow.keras.models import load_model
from ultralytics import YOLO
from collections import Counter

# Load models
rest = load_model('/content/Demo_12_normal_aug.keras')
yolo = YOLO("/content/yolo_130.pt")

# Class names
# class_names = ['Olay_BW_Green', 'Dove_BW_Guava', 'Dove_BW_Purple', 'Olay_BW_Purple_2', 'Olay_BW_Green_2',
#                'Dove_Soap_Brown', 'A_Green', 'Santitas_Yellow', 'Chesters_Red', 'Dove_Soap_Coconut',
#                'Dove_Soap_Yellow', 'A_Yellow', 'Doritos_Red', 'Dove_BW_Blue', 'Garnier_Orange', 'Skin_Purple',
#                'Olay_Soap_Yellow', 'Dove_Soap_Blue_2', 'Dove_BW_Brown', 'Garnier_Brown', 'Olay_BW_Purple',
#                'Dove_Soap_Pink_2', 'Chesters_Orange', 'Vaseline_Brown', 'Dove_Soap_Pink', 'A_Blue',
#                'Dove_BW_Pomo', 'Dove_BW_Pear', 'Olay_Soap_Purple', 'Vaseline_Pink', 'Nivea_Soft', 'Skin_Green',
#                'Vaselin_Green', 'Nivea_White_Lotion', 'Doritos_Purple', 'Chesters_Blue', 'Doritos_Blue',
#                'Garnier_Green', 'Vaselin_Yellow', 'Dove_Soap_Blue', 'Dial_Soap_Yellow', 'Dove_BW_Green',
#                'Dial_Soap_Blue', 'Dove_Soap_Green', 'Nivea_blue_lotion', 'Nivea_Rose', 'Cetaphil']

# class_names = ['Dove_Soap_Brown', 'Dove_Soap_Coconut', 'Dove_Soap_Yellow', 'Doritos_Red', 'Dove_Soap_Blue_2', 'Dove_Soap_Pink_2', 'Dove_Soap_Pink', 'Doritos_Purple', 'Doritos_Blue', 'Dove_Soap_Blue', 'Dove_Soap_Green', 'Nivea_blue_lotion']
# class_names = ["Cocacola","Fanta","Pepsi","Sprite"]

# class_names = ['Tropicana', 'Coca-cola', 'Pepsi', 'Bisleri', 'Sprite', 'Fanta']

# class_names = ['Dove_Soap_Pink', 'Tropicana', 'Coca-cola', 'Pepsi', 'Dove_Soap_Pink_2', 'Dove_Soap_Green', 'Dove_Soap_Blue', 'Bisleri', 'Dove_Soap_Brown', 'Sprite', 'Fanta', 'Dove_Soap_Yellow', 'Dove_Soap_Blue_2', 'Dove_Soap_Coconut']

# class_names = ['Olay_BW_Green', 'Skin_Green', 'Dove_BW_Blue', 'Dove_Soap_Pink', 'Tropicana', 'Dove_BW_Brown', 'Nivea_blue_lotion', 'A_Blue', 'Cetaphil', 'Vaselin_Green', 'Doritos_Purple', 'Chesters_Blue', 'Garnier_Orange', 'Skin_Purple', 'Coca-cola', 'A_Green', 'Dial_Soap_Blue', 'Olay_BW_Purple_2', 'Pepsi', 'Dove_BW_Pomo', 'Garnier_Green', 'Olay_BW_Green_2', 'Doritos_Red', 'Dove_BW_Pear', 'Olay_BW_Purple', 'Chesters_Orange', 'Olay_Soap_Yellow', 'Dove_Soap_Pink_2', 'Santitas_Yellow', 'Garnier_Brown', 'Nivea_White_Lotion', 'Dove_BW_Green', 'Vaseline_Pink', 'Nivea_Soft', 'Vaseline_Brown', 'Vaselin_Yellow', 'A_Yellow', 'Dove_Soap_Green', 'Dove_Soap_Blue', 'Bisleri', 'Doritos_Blue', 'Dove_Soap_Brown', 'Sprite', 'Olay_Soap_Purple', 'Chesters_Red', 'Fanta', 'Nivea_Rose', 'Dove_BW_Guava', 'Dove_Soap_Yellow', 'Dial_Soap_Yellow', 'Dove_BW_Purple', 'Dove_Soap_Blue_2', 'Dove_Soap_Coconut']

class_names = ['Tropicana', 'Doritos_Purple', 'Garnier_Orange', 'Coca-cola', 'Pepsi', 'Garnier_Green', 'Doritos_Red', 'Garnier_Brown', 'Bisleri', 'Doritos_Blue', 'Sprite', 'Fanta']
# Normalize image
def normalize_image(img):
    img = img.astype('float32')
    img -= np.mean(img)
    img /= np.std(img)
    return img
```

```python
import matplotlib.pyplot as plt
from io import BytesIO

def generate_donut_chart(own_count, competitor_count):
    labels = ['Own Products', 'Competitor Products']
    sizes = [own_count, competitor_count]
    colors = ['#4CAF50', '#F44336']  # Green for own, Red for competitor

    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90,
           colors=colors, pctdistance=0.85)
    centre_circle = plt.Circle((0, 0), 0.70, fc='white')
    fig.gca().add_artist(centre_circle)
    ax.axis('equal')
    plt.title("Own vs Competitor Products")

    buf = BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight')
    plt.close(fig)
    buf.seek(0)
    return Image.open(buf)

# Main detection function
def detect_and_count(image: Image.Image):
    image = image.convert("RGB")
    image_np = np.array(image)
    draw_image = image.convert("RGBA")
    overlay = Image.new("RGBA", draw_image.size, (255, 255, 255, 0))
    overlay_draw = ImageDraw.Draw(overlay)

    results = yolo(image)
    boxes = results[0].boxes.xyxy

    own_count = 0
    competitor_count = 0
    final_products = []
    total_products = 0

    for box in boxes:
        x1, y1, x2, y2 = map(int, box[:4])
        crop = image_np[y1:y2, x1:x2]
        cropped_img = Image.fromarray(crop).resize((256, 256)).convert("RGB")
        img_array = np.array(cropped_img)
        img = normalize_image(img_array)
        img = np.expand_dims(img, axis=0)

        class_probabilities = rest.predict(img)[0]
        # print(class_probabilities)
        predicted_index = np.argmax(class_probabilities)
        confidence = class_probabilities[predicted_index]
        print(confidence)
        if confidence < 0.7:
            predicted_class = "Competitor Product"
            fill_color = (255, 0, 0, 160)
            competitor_count += 1
            total_products += 1
        else:
            predicted_class = class_names[predicted_index]
            fill_color = (0, 128, 0, 160)
            own_count += 1
            total_products += 1

        print(predicted_class)
        final_products.append(predicted_class)

        # Fill bounding box with transparent color
        overlay_draw.rectangle([x1, y1, x2, y2], fill=fill_color)
        # overlay_draw.text((x1, max(0, y1 - 12)), predicted_class, fill=(0, 0, 0, 255))

    # Combine overlay with base image
    draw_image = Image.alpha_composite(draw_image, overlay).convert("RGB")

    # Count summary
    counter = Counter(final_products)
    count_summary = "\n".join(f"{label} : {count}" for label, count in counter.items()) or "No products detected"


    # Generate donut chart
    donut_chart = generate_donut_chart(own_count, competitor_count)

    return draw_image, donut_chart, count_summary


app = gr.Interface(
    fn=detect_and_count,
    inputs=gr.Image(type="pil", label="Upload Product Shelf"),
    outputs=[
        gr.Image(type="pil", label="Detected Products"),
        gr.Image(type="pil", label="Own vs Competitor Donut Chart"),
        gr.Textbox(label="Product Summary")
    ],
    title="Share of Shelf Analysis",
    description="Upload a shelf image"
)

app.launch(debug=True, share=True)
```
